## Names and objects

As mentioned earlier, everything is an object in Python, and names can be attached to these so we can refer back to them. To illustrate further the implications of this, take this example:

```python
# say this is my weight in kilograms
weight_kg = 55
```

```python
# I'll calculate my weight in pounds
weight_lb = 2.2 * weight_kg

print(weight_lb)
```

Apparently I got it wrong, my weight is not that low! Let me try again:

```python
# say this is my weight in kilograms
weight_kg = 70
```

```python
# now we should be good
print(weight_lb)
```

What's happening here?

`weight_lb` is just a name attached to a value, in this case a `float`. It knows nothing about how that value came to be. In order to update `weight_lb`, we must re-run the code we calculated it with given the new value of `weight_kg`:

```python
# I'll calculate my weight in pounds
weight_lb = 2.2 * weight_kg

print(weight_lb)
```

So now the value is updated. This behavior is different than, say, spreadsheet cells work, in which a formula given to a cell automatically shows updated values when the cells the formula refers to are updated.

## Repetition with loops 

Say we have a word:

```python
word = 'bird'
```

and the word is 'bird'. We want to print each letter of the word to a separate line. We could do this with:

```python
print(word[0])
print(word[1])
print(word[2])
print(word[3])
```

But this doesn't scale. What if our word was something more substantial, like:

```python
word = 'supercalifragilisticexpialidocious'
```

One thing that computers do remarkably well that humans do not: the same thing, repeatedly, over and over again. We should use a loop:

```python
for letter in word:
    print(letter)
```

**Note**: Python is indentation-sensitive! The body of the for-loop is one indentation level, or four spaces, from the level of the `for` statement itself. Statements that follow the loop should be at the original indentation level.

### Breakout: Write a loop that counts the number of characters in a word.

Answer:

```python
count = 0
for letter in word:
    count += 1

print(count)
```

But a better solution is to use the builtin function `len`:

```python
len(word)
```

## Making decisions

Often it's necessary to write code that does different things depending on some set of conditions. That is, the code is set up to make decisions using **conditionals**. Say we have some number:

```python
num = 37
```

We can ask whether that number is greater than 100 with a conditional like this:

```python
if num > 100:
    print('greater')
else:
    print('not greater')
```

We could add some more sophistication to the conditional by checking for equality:

```python
if num > 100:
    print('greater')
elif num == 100:
    print('equal!')
else:
    print('not greater')
```

...and we could test it:

```python
num = 100
```

```python
if num > 100:
    print('greater')
elif num == 100:
    print('equal!')
else:
    print('not greater')
```

Conditionals can include boolean operators such as ``and``:

```python
if (1 > 0) and (-1 > 0):
    print('both parts are true')
else:
    print('one part is not true')
```

A "truth table" for ``and``, given two boolean values ``A`` and ``B``, each either ``True`` or ``False`` gives:

| `A`   | `B`   | `A and B` |
| ----- | ----- | --------- |
| True  | True  | True      |
| True  | False | False     |
| False | True  | False     |
| False | False | False     |

There's also `or`:

```python
if (1 > 0) or (-1 > 0):
    print('at least one part is true')
else:
    print('no part is true')
```

| `A`   | `B`   | `A or B` |
| ----- | ----- | -------- |
| True  | True  | True     |
| True  | False | True     |
| False | True  | True     |
| False | False | False    |

And `not`:

```python
if (1 > 0) and not (-1 > 0):
    print('both parts are true')
else:
    print('one part is not true')
```

| `A`   | `not A` |
| ----- | ------- |
| True  | False   |
| False | True    |

## Functions: modular, reusable, and testable pieces of code

Although it's totally natural to write blocks of code when prototyping or when exploring data, the trouble with this is that applying the same code elsewhere, such as later in this same notebook, requires copying and pasting.

This is a problem because if you discover an issue with the code block or add additional functionality, you must manually add this to each instance you created. Not good. Instead, we should make a reused bit of code into a function.

We've already been using functions. One example is the python built-in ``len``:

```python
stuff
```

```python
len(stuff)
```

```python
# run this to view the docstring
len?
```

Let's write a simple function to show the general form; the function ``smaller`` tests whether the input value is smaller than ``10``, then returns either ``True`` or ``False`` depending:

```python
# def is followed by the name of the new function, then
# its arguments in parentheses
def smaller(somenumber):
    if somenumber < 10:    # mind the indentation levels!
        out = True
    else:
        out = False
    
    return out             # return gives the value to be output by the function
```

```python
smaller(157)
```

```python
smaller(7)
```

```python
smaller(10)
```

In fact, for such a simple function it can be condensed greatly:

```python
def smaller(somenumber):        
    return somenumber < 10
```

Writing functions make it easier to write *better* code as well as reusable code. *Better* in that it is easier to test, and therefore more likely to actually do what you want it to do. Functions can (and should!) be written to do individual tasks (like bash tools) so they can be used as building blocks for larger programs. As an example, let's write a function that calculates the temperature in Kelvin from a temperature in Fahrenheit:

```python
def fahr_to_kelvin(temp):
    return ((temp - 32) * (5/9)) + 273.15
```

Now, let's do some quick sanity tests:

```python
fahr_to_kelvin(32)
```

The freezing point of water looks good!

```python
fahr_to_kelvin(212)
```

Good!

What about a Kelvin to Celsius converter?

```python
def kelvin_to_celsius(temp):
    return temp - 273.15
```

```python
kelvin_to_celsius(273.15)
```

```python
kelvin_to_celsius(373.15)
```

The sanity checks work out.

### Breakout: Write a function that converts fahrenheit to celsius.

Answer:

```python
def fahrenheit_to_celsius(temp):
    """Convert fahrenheit to celsius temperatures.
    
    Parameters
    ----------
    temp : float
        Temperature in fahrenheit
        
    Returns
    -------
    temp : float
        Temperature in celsius
    
    """
    temp = fahr_to_kelvin(temp)
    temp = kelvin_to_celsius(temp)
    return temp
```

note that we can make this even simpler:

```python
def fahrenheit_to_celsius(temp):
    """Convert fahrenheit to celsius temperatures.
    
    Parameters
    ----------
    temp : float
        Temperature in fahrenheit
        
    Returns
    -------
    temp : float
        Temperature in celsius
    
    """
    return kelvin_to_celsius(fahr_to_kelvin(temp))
```

Note how we *composed* two functions we previously wrote together instead of copying and pasting their contents. Now, if we later find a problem in how we are converting fahrenheit to kelvin, we don't need to fix it in multiple places. There is precisely one place that needs fixing.

We have also included a docstring that specifies the inputs and outputs of the function. This is important, since 1) the person most likely to read your code is your future self, and 2) it allows you to wield the function without knowing how it goes from its inputs to its outputs.
